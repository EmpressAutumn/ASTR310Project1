```python
import numpy as np
import matplotlib.pyplot as plt
from astropy.io import fits
```

```python
bias_set = []

for i in range(7):
    string = '/Users/kushpatel/Downloads/NGC6946_000' + str(i) + '-ha.fits'
    terpimagename = fits.open(string)
    temp_data = terpimagename[0].data
    bias_set.append(temp_data)
```

```python
def median_combine(data):
    final_list = []
    for j in range(len(data[0])):
        temp_list = []
        for k in range(len(data[0][0])):
                median_combiner = []
                for i in range(len(data)):
                    median_combiner.append(data[i][j][k])
                sorted_combiner = sorted(median_combiner)
                temp_list.append(np.median(np.array(sorted_combiner)))
        final_list.append(temp_list)
    return final_list
```

```python
bias_master = median_combine(bias_set)
```

```python
hdu = fits.PrimaryHDU(bias_master)
hdu.writeto('/Users/kushpatel/Downloads/bias_master.fits', overwrite=True)
```
