```python
import numpy as np
```

```python
def median_combine(data(list)):
    final_list = []
    for j in range(len(data[0])):
        temp_list = []
        for k in range(len(data[0][0])):
                median_combiner = []
                for i in range(len(data)):
                    median_combiner.append(data[i][j][k])
                sorted_combiner = sorted(median_combiner)
                temp_list(np.median(sorted_combiner))
        final_list.append(temp_list)
    return final_list
```
